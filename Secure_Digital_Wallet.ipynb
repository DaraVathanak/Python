```python
class DigitalWallet:
    def __init__(self,balance,pin_code):
        self.__balance = balance
        self.__pin_code = pin_code
    def __verify(self,pin):
        return self.__pin_code == pin
    def deposit(self,amount):
        if amount <= 0:
            print("Error")
            return
        self.__balance += amount
        print(f"Deposited: {amount:.2f}")
    def withdraw (self,amount,pin):
        if not self.__verify(pin):
            print("Incorrect PIN")
            return
        if amount <= 0:
            print("Error")
            return
        if amount > self.__balance:
            print("Not enough")
            return
        self.__balance -= amount
        print(f"Withdraw: {amount:.2f}")
    def get_balance(self, pin):
        if not self.__verify(pin):
            print("Incorrect PIN")
            return 
        print(f"Your Balance: ${self.__balance:.2f}")
        return self.__balance
    @staticmethod
    def menu():
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")
pin    = int(input("Set your PIN: ").strip())
wallet = DigitalWallet(pin_code=pin, balance=1000)
while True:
    DigitalWallet.menu()
    option = input("Choose: ").strip()
    if option == "1":
        amount = float(input("Enter deposit amount: $").strip())
        wallet.deposit(amount)
    elif option == "2":
        amount = float(input("Enter withdraw amount: $").strip())
        pin    = int(input("Enter PIN: ").strip())
        wallet.withdraw(amount, pin)
    elif option == "3":
        pin = int(input("Enter PIN: ").strip())
        wallet.get_balance(pin)
    elif option == "4":
        break
    else:
        print("Invalid")
```
