# The Climate Sensor Grid Analyzer

##### The Scenario: You are a climate data scientist. You have been handed a massive dataset from a satellite grid. To load the .npy file use np.load(“...”). The data is a 3D NumPy array structured as (Latitude, Longitude, Days), and the values inside represent the daily peak temperature in Celsius. You need to clean, analyze, and transform this data using only NumPy operations.

##### Creating Arrays & Attributes: You load the provided climate dataset. You must inspect .shape, .ndim, and .dtype to understand how the geographical coordinates and the 31 days of the month are structured. Because you will be calculating temperature anomalies later, you also need to create a blank "baseline grid" of the exact same geographical dimensions using np.zeros().

```python
import numpy as np
data = np.load("climate_data.npy")
print("Shape:", data.shape)
print("Dimensions:", data.ndim)
print("Data type:", data.dtype)
baseline = np.zeros(data.shape[:2])
```

### Indexing, Slicing & Masking: 

##### Slicing: The dataset covers the whole globe, but you only want to analyze a specific drought-prone region. Crop the data to specific latitude and longitude coordinates (e.g., climate_data[20:50, 70:120, :]).

```python
import numpy as np
data = np.load("climate_data.npy")
region = data[20:50, 70:120, :]
print(region.shape)
```

##### Indexing: Isolate a "snapshot" of the entire grid for exactly Day 15 to see a cross-section of global temperatures on that specific day.

```python
import numpy as np
data = np.load("climate_data.npy")
day15 = data[:, :, 14]
print(day15.shape)
```

##### Masking: Data cleaning is required. A sensor glitch caused several locations to record impossible temperatures of -99°C. Create a boolean mask to find all instances of -99 and replace them with np.nan (Not a Number) so they don't skew your later calculations.

```python
import numpy as np
data = np.load("climate_data.npy")
mask = data == -99.0
data[mask] = np.nan
print(np.sum(mask))
```

### Vectorization, Broadcasting & Ufuncs: 

##### Broadcasting: Your American stakeholders need the data in Fahrenheit. Convert the entire 3D array from Celsius to Fahrenheit in a single mathematical step ((data * 9/5) + 32) without writing a single for loop.

```python
import numpy as np
data = np.load("climate_data.npy").astype(float)
data[data == -99.0] = np.nan
data_f = (data * 9/5) + 32
print(data_f[0, 0, :])
```

#####  Ufuncs: To model a specific heat-stress index, apply a non-linear transformation to the temperature values using np.exp() and np.log() across the entire grid simultaneously.

```python
import numpy as np
data = np.load("climate_data.npy").astype(float)
data[data == -99.0] = np.nan
stress = np.log1p(np.exp(data / 40.0))
print(stress[0, 0, :5])
```

###  Math & Stats:


##### Find the absolute highest and lowest valid temperatures recorded anywhere on the grid during the month (np.nanmax, np.nanmin).

```python
import numpy as np
data = np.load("climate_data.npy").astype(float)
data[data == -99.0] = np.nan
max_temp = np.nanmax(data)
min_temp = np.nanmin(data)
print("Max temperature:", max_temp)
print("Min temperature:", min_temp)
```

##### Calculate the average temperature for each specific sensor location across the entire month (using np.nanmean across the "Days" axis). This will collapse your 3D array into a 2D map showing the baseline climate for each coordinate!

```python
import numpy as np
data = np.load("climate_data.npy").astype(float)
data[data == -99.0] = np.nan
baseline_map = np.nanmean(data, axis=2)
print(baseline_map.shape)
print(baseline_map[0, 0:5])
```
