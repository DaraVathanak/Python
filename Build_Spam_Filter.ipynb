# Build a Spam Filter
##### Before modern Artificial Intelligence could read emails to determine if they were spam, early spam filters used rule-based programming. They would scan an email and count how many spammy words were used. If the percentage of spam words was too high, the email was blocked.

##### Your task is to build a program that analyzes a string of text, counts the total words, counts the trigger words, and calculates a final Spam Score.

##### The Trigger Words: ["urgent", "winner", "free", "money", "click", "guarantee", "act", "prize"]


##### Dataset

##### email_1 = """
##### URGENT: You are a winner! Click here to claim your free money guarantee.This is not a scam, it is 100% free. Act now to get your prize!\
##### """

##### email_2 = """
Hey team, just a quick reminder about the meeting tomorrow at 10 AM.
Please bring your project updates. Thanks!
"""


email_3 = """
I am writing to see if you have any free time this weekend.
I owe you money for lunch, so let me buy you a pizza!
"""

```python
trigger_word = ["urgent", "winner", "free", "money", "click", "guarantee", "act", "prize"]
email_1 = """
URGENT: You are a winner! Click here to claim your free money guarantee.
This is not a scam, it is 100% free. Act now to get your prize!
"""
email_2 = """
Hey team, just a quick reminder about the meeting tomorrow at 10 AM.
Please bring your project updates. Thanks!
"""
email_3 = """
I am writing to see if you have any free time this weekend.
I owe you money for lunch, so let me buy you a pizza!
"""
emails = [("Email1", email_1),("Email2", email_2),("Email3", email_3)]
for label, email in emails:
    words = email.lower().split()
    words = [w.strip(".,!?:;\"'") for w in words]
    total_words = len(words)
    found_triggers = [w for w in words if w in trigger_word]
    trigger_count = len(found_triggers)
    spam_score = (trigger_count/total_words)*100 if total_words > 0 else 0.0
    if spam_score >= 10:
        print("SPAM")
    elif spam_score >= 5:
        print("SUSPICIOUS")
    else:
        print("CLEAN")
```
