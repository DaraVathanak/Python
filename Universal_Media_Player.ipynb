```python
class Media:
    def play(self):
        pass
class Song(Media):
    def play(self):
        print("Playing audio...")
class Movie(Media):
    def play(self):
        print("Playing video at 1080p...")
class Podcast(Media):
    def play(self):
        print("Playing podcast")
def play_all(media_list):
    for media in media_list:
        media.play()
song = Song()
movie = Movie()
podcast = Podcast()
playlist = [song, movie, podcast]
play_all(playlist)
```
