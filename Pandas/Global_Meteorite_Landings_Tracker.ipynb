### Indexing and Selection: 
##### Column/Label/Integer: Use dictionary-style selection to grab the fall_status column (whether it was seen falling or just found later). Use .iloc[] to extract the top 5 heaviest meteorites based on their sorted row position.

```python
df.loc[:,['fall_status']]
```

```python
heaviest = df.sort_values(by="mass_grams",ascending=False)
heaviest.iloc[0:5]
```

##### Filtering: Apply multiple boolean conditions using & and | to find meteorites that are both greater than 50,000 grams in mass and were discovered in a specific region (e.g., "Antarctica" or "Sahara").

```python
filters = (df["mass_grams"] > 50000) & (
            (df["region"] == "Antarctica") | (df["region"] == "Sahara")
         )
filtered = df[filters]
filtered
```

##### Data I/O: You start by loading the raw data using pd.read_csv(). Later, you will export your filtered findings using .to_excel() or .to_json() for a museum curation team to use for an upcoming exhibit.

```python
filtered.to_json("meteorites.json", orient="records")
print("Success")
```
