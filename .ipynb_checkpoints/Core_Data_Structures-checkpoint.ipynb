# Memory-Based To-Do List:
##### The Goal: An application that lets a user add a task, view all tasks, remove a task, and exit. Store the tasks in a Python List.
##### Concepts Tested: List methods (.append(), .pop()), indexing, iterating through lists with for loops.

```python
task = []
print("Type 'Add' to add the task")
print("Type 'Remove' to delete task")
print("Type 'View' to see all task")
print("Type 'stop' to exit")
while True:
    user_input = input("Choose Add,Remove,View:").lower().strip()
    if user_input == "stop":
        break
    if user_input == "add":
        Add = input("Enter the task:").strip()
        if task:
            position = input("Insert position or enter to add to the end:").strip()
            if position == "":
                task.append(Add)
            elif position.isdigit() and 1 <= int(position) <= len(task)+1:
                task.insert(int(position)-1,Add)
            else:
                task.append(Add)
        else:
            task.append(Add)
    elif user_input == "remove":
        i = 1 
        for t in task:
            print(f"{i}. {t}")
            i+=1
        Remove = input("Type position to remove:").strip()
        if Remove.isdigit() and 1 <= int(Remove) <= len(task):
            removed = task.pop(int(Remove)-1)
        else:
            print("Error")
    elif user_input == "view":
        i = 1
        for t in task:
            print(f"{i}. {t}")
            i+=1
    else:
        print("Invalid")
```

# Interactive Flashcard / Quiz App:
##### The Goal: Create a dictionary where the keys are questions (e.g., countries) and values are answers (e.g., capitals). Loop through the dictionary, ask the user the question, check their answer, and calculate a final score.
##### Concepts Tested: Dictionaries, key-value pairs, iteration, score tracking.

```python
country = {
    "Cambodia" : "phnom penh",
    "US" : "new york",
    "UK" : "london",
    "French" : "paris"
}
user_score = 0
for key, value in country.items():
    user_answer = input(f"The Capital of {key} is:").lower().strip()
    if user_answer == value:
        print("Your answer Correct")
        user_score += 1
    else:
        print(f"Your answer Wrong the answer is: {value}")
print(f"Final Score: {user_score}")
```

# The Shopping Cart Analyzer:
##### The Goal: Ask the user to input the prices of items they are adding to their cart, one by one. Store these prices in a Python List. The loop stops when the user types 'done'. Then, calculate and print a receipt showing the total cost, the most expensive item in the cart, and the average item price.
##### Concepts Tested: while loops for input collection, list methods (.append()), built-in list functions (sum(), max(), len()), and data type casting (handling the mix of string commands and float prices).

```python
cart = []
print("Type 'done' when finnished")
while True:
    enter = input("Enter price:")
    if enter == "done":
        break
    try:
        price = float(enter)
        cart.append(price)
        print(f"Added: ${price:.2f}")
    except ValueError:
        print("Invalid input number only")
if len(cart) == 0:
    print("Your cart empty")
else:
    total = sum(cart)
    expensive = max(cart)
    average = total/len(cart)
    print(f"Total: {total:.2f}")
    print(f"Most Expensive: {expensive:.2f}")
    print(f"Average: {average:.2f}")
```

# The Coffee Shop Inventory Manager:
##### The Goal: Create a dictionary pre-filled with cafe supplies and their quantities (e.g., {'coffee beans': 50, 'cups': 100, 'syrup': 15}). Create a loop that asks the user if they want to "sell" or "restock" an item. Update the dictionary values based on their input and print the current inventory levels. Prevent the user from selling more items than they have in stock!
##### Concepts Tested: Accessing and updating dictionary values, checking if a key exists using the in operator, combining if/elif logic with dictionary lookups, and managing state.

```python
inventory = {"coffee beans": 50, "cups": 100, "syrup": 15}
print("Options: Sell, Restock, View, Exit")
while True:
    user_input = input("Choose an option:").lower().strip()
    if user_input == "exit":
        break
    elif user_input == "sell":
        item = input("Item name:").lower().strip()
        if item in inventory:
            qty = int(input("Enter the amount:"))
            if qty > inventory[item]:
                print("Not enough stock")
            else:
                inventory[item] -= qty
                print("Sell Success")
        else:
            print("Item not found")
    elif user_input == "restock":
        item = input("Item name:").lower().strip()
        if item in inventory:
            qty = int(input("Enter the amount:"))
            inventory[item] += qty
        else:
            print("Item not found")
    elif user_input == "view":
        for key, value in inventory.items():
            print(f"{key}:{value}")
    else:
        print("Invalid option")
```
