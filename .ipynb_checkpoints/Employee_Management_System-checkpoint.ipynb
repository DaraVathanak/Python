```python
class Employee:
    def __init__(self,name,base_salary):
        self.name = name
        self.base_salary = base_salary
    def calculate_pay(self):
        return self.base_salary
class Manager(Employee):
    def __init__(self,name,base_salary,bonus):
        super().__init__(name,base_salary)
        self.bonus = bonus
    def calculate_pay(self):
        return self.base_salary + self.bonus
class Developer(Employee):SS
    def __init__(self,name,base_salary,programming_language):
        super().__init__(name,base_salary)
        self.programming_language = programming_language
    def write_code(self):
        print(f"{self.name} is writing code in {self.programming_language}")
manager = Manager("John",5000,1500)
developer = Developer("Alice",5000, "Python")
print(f"{manager.name} salary:  ${manager.calculate_pay()}")
print(f"{developer.name} salary:  ${developer.calculate_pay()}")
developer.write_code()
```
