# Modular Unit Converter:
##### The Goal: A program that converts temperatures (Celsius to Fahrenheit), distances (Miles to Kilometers), and weights(Grams to Kilograms).
##### The Catch: Every single conversion must be its own independent function (e.g., def c_to_f(celsius):). There should be a main menu function that calls the others based on user choice.
##### Concepts Tested: Defining functions, passing arguments, return statements, variable scope.

```python
def c_to_f(celsius):
    result = (celsius*9/5)+32
    return result
def miles_to_km(miles):
    result = miles*1.60934
    return result
def g_to_kg(g):
    result = g/1000
    return result
def main():
    print("1. Celsius to Fahrenheit")
    print("2. Miles to Kilometers")
    print("3. Grams to Kilograms")
    print("4. Exit")
while True:
    main()
    option = input("Choose:").strip()
    if option == "4":
        break
    elif option == "1":
        value = float(input("Enter Celsius:")).strip()
        result = c_to_f(value)
        print(f"{value:.2f} = {result:.2f}")
    elif option == "2":
        value = float(input("Enter Miles:")).strip()
        result = miles_to_km(value)
        print(f"{value:.2f} = {result:.2f}")
    elif option == "3":
        value = float(input("Enter Grams:")).strip()
        result = g_to_kg(value)
        print(f"{value:.2f} = {result:.2f}")
    else:
        print("Invalid Option")
```

# Password Generator:
##### The Goal: A function that takes an integer (length) and boolean flags (use_numbers, use_symbols) and returns a randomized, secure password.
##### Concepts Tested: Default arguments, keyword arguments, combining lists/strings, returning data.

```python
import random
import string
letter = string.ascii_letters
number = string.digits
symbol = string.punctuation
def generate_pass(length, use_number=False,use_symbol=False):
    pas = letter
    if use_number:
        pas+=number
    if use_symbol:
        pas+=symbol
    if not pas:
        return "Error"
    password = ""
    for _ in range(length):
        password += random.choice(pas)
    return password
while True:
    length = input("How long password be:")
    length = int(length.strip())
    if length <= 0 :
        print("Error")
        continue
    add_number = input("Want to use number(Y/N):").lower().strip()
    use_number = True if add_number == "y" else False
    add_symbol = input("Want to use symbol(Y/N):").lower().strip()
    use_symbol = True if add_symbol == "y" else False
    password = generate_pass(length,use_number=use_number,use_symbol=use_symbol)
    print(f"Your Password: {password}")
    ask = input("Want to generate more(Y/N):").lower().strip()
    if ask != "y":
        break
```

# The Text Analyzer Suite:
##### The Goal: Ask the user to input a long sentence or paragraph. Build a program that outputs statistics about their text.
##### The Catch: You must use separate functions for every analysis: count_words(text), count_vowels(text), and find_longest_word(text). Each function must take the text as an argument and return the result to be printed by the main program.
##### Concepts Tested: String manipulation, passing strings as arguments, returning specific values (integers and strings) from functions, and keeping logic strictly separated.

```python
user_enter = input("Enter any sentence or paragraph:").strip()
def count_word(text):
    words = text.lower().split()
    words = [w.strip(".,!?:;\"'") for w in words]
    return len(words)
def count_vowel(text):
    vowel = {"a", "e", "i", "o", "u"}
    count = 0
    for char in text.lower():
        if char in vowel:
            count += 1
    return count   
def find_longest_word(text):
    words = text.lower().split()
    words = [w.strip(".,!?:;\"'") for w in words]
    if not words:
        print("Error")
    longest = ""
    for word in words:
        if len(word) > len(longest):
            longest = word
    return longest
word_count  = count_word(user_enter)
vowel_count = count_vowel(user_enter)
longest_word  = find_longest_word(user_enter)
print(f"Amount of words: {word_count}")
print(f"Amount of vowels: {vowel_count}")
print(f"The longest word: {longest_word}")
```
