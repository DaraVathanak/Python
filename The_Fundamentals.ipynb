```python
user_bill = float(input("The total cost:"))
user_tip = float(input("How much tip percentage you want to give:"))
tip = user_bill * (user_tip / 100)
total = tip + user_bill
print(f"Tip amount:   ${tip:.2f}")
print(f"Total amount: ${total:.2f}")
```

```python
user_distance = float(input("Input the Distance in Km:"))
user_KPL = float(input("Input the Kilometers Per Liter:"))
fuel_price = int(input("Input price Fuel Per Liter: ៛"))
fuel = user_distance / user_KPL
total_cost = fuel * fuel_price
print(f"Distance: {user_distance:.2f} Km")
print(f"Fuel needed: {fuel:.2f} L")
print(f"Total cost of trip: ៛{total_cost:.2f}")
```
