```python
class Pet:
    def __init__(self,name,hunger,energy,happiness):
        self.name = name
        self.hunger = hunger
        self.energy = energy
        self.happiness = happiness
    def feed(self):
        self.hunger -= 1
        self.happiness += 1
        print(f"Feeding {self.name}")
    def play(self):
        self.happiness += 1
        self.energy -= 1
        print(f"{self.name} is playing")
    def sleep(self):
        self.energy += 1
        self.hunger += 1
        print(f"{self.name} is sleeping")
    def status(self):
        print(f"{self.name}")
        print(f"Hunger: {self.hunger}")
        print(f"Energy: {self.energy}")
        print(f"Happiness: {self.happiness}")
    def menu():
        print("1. Feed")
        print("2. Play")
        print("3. Sleep")
        print("4. Exit")
pet = Pet("Corgi",hunger=5,energy=5,happiness=5)
pet.status()
while True:
    Pet.menu()
    option = input("Choose:").strip()
    if option == "1":
        pet.feed()
        print(f"Hunger: {pet.hunger}/10")
        print(f"Happiness: {pet.happiness}/10")
    elif option == "2":
        pet.play()
        print(f"Happiness: {pet.happiness}/10")
        print(f"Energy: {pet.energy}/10")
    elif option == "3":
        pet.sleep()
        print(f"Energy: {pet.energy}/10")
        print(f"Hunger: {pet.hunger}/10")
    elif option == "4":
        break
    else:
        print("Invalid")
        
    
```
