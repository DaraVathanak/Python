# Number Guessing Game:
##### The Goal: The program randomly selects a number between 1 and 100. The user guesses the number, and the program replies with "Too high!", "Too low!", or "You got it!". The game loops until the user guesses correctly.
##### Concepts Tested: while loops, if/elif/else statements, comparison operators, importing the basic random module.

```python
import random
number = random.randint(1, 100) 
print("Guess the number between 1 and 100")
while True:
    guess = int(input("Enter your number to guess:"))
    if guess < number:
        print("Higher")
    elif guess > number:
        print("Lower")
    else:
        print(f"Congratulations, you guess the correct number is {number}")
        break
```

# Rock, Paper, Scissors:
##### The Goal: A continuous loop where the user plays Rock, Paper, Scissors against the computer. Keep track of the score until the user types "quit".
##### Concepts Tested: Infinite loops with break conditions, complex if/elif logic, string normalization (e.g., using .lower()).

```python
import random
choices = ["rock","paper","scissors"]
user_score = 0
computer_score = 0
ties = 0
print("Type 'quit' to stop and see final score")
while True:
    user_input = input("Enter rock, paper, or scissors:").lower()
    if user_input == "quit":
        print(f"Your Score: {user_score}")
        print(f"Computer Score: {computer_score}")
        print(f"Ties: {ties}")
        break
    if user_input not in choices:
        print("Invalid input type only rock, paper, or scissors")
        continue
    computer_input = random.choice(choices)
    print(f"Computer: {computer_input}")

    if user_input == computer_input:
        print("Tie")
        ties += 1
    elif (user_input == "rock" and computer_input == "scissors") or (user_input == "paper" and computer_input == "rock") or (user_input == "scissors" and computer_input == "paper"):
        print("You Win")
        user_score += 1
    else:
        print("Computer Win")
        computer_score += 1
```

# The Virtual Barista (Simple Point-of-Sale)
##### The Goal: Display a text menu of three coffee shop drinks with different prices. Create a continuous loop that asks the user what they want to order. Add the price of their chosen drink to a running total. If they type an invalid drink, tell them it's not on the menu. The loop should continue until the user types "checkout", at which point the program prints their final total.
##### Concepts Tested: Infinite while loops, break conditions, accumulating totals (+= operator), complex if/elif/else logic, and string normalization (using .lower() to handle if the user types "Checkout" or "CHECKOUT").

```python
menu = {
    "espresso":   2.50,
    "ice_latte":      4.00,
    "cappuccino": 3.50
}
total = 0
order_amount = 0

print("MENU")
print(f"Espresso: ${menu['espresso']:.2f}")
print(f"Ice_Latte: ${menu['ice_latte']:.2f}")
print(f"Cappuccino: ${menu['cappuccino']:.2f}")
print("Type 'checkout' when done")
while True:
    order = input("What would you want to order:").lower()
    if order == "checkout":
        print(f"Total Item: {order_amount}")
        print(f"Total amount: {total:.2f}")
        if order_amount == 0:
            print("You didn't order yet")
        else:
            print("Thank you")
        break
    elif order == "espresso":
        total += menu["espresso"]
        order_amount += 1
        print(f"Esprsso added: ${menu['espresso']:.2f}")
        print(f"Total: ${total:.2f}")
    elif order == "ice_latte":
        total += menu["ice_latte"]
        order_amount += 1
        print(f"Ice_latte added: ${menu['ice_latte']:.2f}")
        print(f"Total: ${total:.2f}")
    elif order == "cappuccino":
        total += menu["cappuccino"]
        order_amount += 1
        print(f"Cappuccino added: ${menu['cappuccino']:.2f}")
        print(f"Total: ${total:.2f}")
    else:
        print("The order not on the order, Please choose in the menu")
    
```

# The Mental Math Challenger:
##### The Goal: The program generates two random numbers between 1 and 12 and asks the user to multiply them together (e.g., "What is 7 x 8?"). If the user gets it right, they get a point, and the program asks another question. If they get it wrong, the program tells them the correct answer and ends the game, displaying their final streak/score.
##### Concepts Tested: Importing the random module, tracking score variables, while loops dependent on a boolean condition (e.g., while is_correct:), if/else statements, and type casting (converting the user's answer to an int to compare it to the actual math result).

```python
# import random
user_score = 0
is_correct = True
print("Mutiply the two number")
while is_correct:
    num1 = random.randint(1,12)
    num2 = random.randint(1,12)
    answer = num1 * num2
    print(f"What is {num1} X {num2} = ?")
    user_answer = int(input("Your answer:"))
    if user_answer == answer:
        user_score += 1 
        print("Your answer correct")
    else:
        print(f"Your answer wrong, The correct answer is {answer}")
        is_correct = False
print(f"Your total score : {user_score}")
```
